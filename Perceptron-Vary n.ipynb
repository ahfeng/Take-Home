```python
import random
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
num_examples=20
```

```python
a = random.randint(-10,10)
b = random.randint(-10,10)
x1 = [0]*num_examples
x2 = [0]*num_examples
y = [-1]*num_examples
w = [1]*3
incorrect = []
print("Target function x2 = ",a,"x1 -",b)
for i in range(num_examples):
    x1[i] = random.randint(-50,50)
    x2[i] = random.randint(-50,50)
    print((x1[i],x2[i]))
    if(x2[i]-a*x1[i]-b>=0):
        y[i]=1
```

```python
fig = plt.figure()
ax = plt.axes()
plt.xlim(-50, 50)
plt.ylim(-50, 50)
x = np.linspace(-50, 50, 1000)
plt.plot(x,a*x+b)
for i in range(num_examples):
    if(y[i]==1):
        plt.plot(x1[i],x2[i],'o',color='blue')
    else:
        plt.plot(x1[i],x2[i],'o',color='red')
plt.plot(x,-w[1]/w[2]*x-w[0]/w[2])
```

```python
def evaluate(x1,x2,w,y,incorrect):
    incorrect = []
    for i in range(num_examples):
        output = np.sign(w[0]+w[1]*x1[i]+w[2]*x2[i]);
        if(y[i]!=output):
            incorrect.append(i)
    return incorrect
```

```python
incorrect = evaluate(x1,x2,w,y,incorrect)
count = 0
plt.xlim(-50, 50)
plt.ylim(-50, 50)
while(len(incorrect)>0):
    index = random.choice(incorrect)
    w[0] = w[0] + y[index]
    w[1] = w[1] + y[index]*x1[index]
    w[2] = w[2] + y[index]*x2[index]
    incorrect = evaluate(x1,x2,w,y,incorrect)
    count=count+1
    print("After ",count, "iterations")
    if(w[2]!=0):
        plt.plot(x,-w[1]/w[2]*x-w[0]/w[2])
```

```python
print("Final w:",w)
print("Final hypothesis: x2 = ",-w[1]/w[2],"x1 -",w[0]/w[2])
plt.xlim(-50, 50)
plt.ylim(-50, 50)
plt.plot(x,a*x+b)
for i in range(num_examples):
    if(y[i]==1):
        plt.plot(x1[i],x2[i],'o',color='blue')
    else:
        plt.plot(x1[i],x2[i],'o',color='red')
plt.plot(x,-w[1]/w[2]*x-w[0]/w[2])
```
