```python
import random
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
a = random.randint(-10,10)
b = random.randint(-10,10)
x1 = [0]*100
x2 = [0]*100
y = [-1]*100
x1test = [0]*10000
x2test = [0]*10000
ytest = [-1]*10000
w = [1]*3
nu = 100
incorrect = []
print("Target function x2 = ",a,"x1 -",b)
for i in range(100):
    x1[i] = random.randint(-50,50)
    x2[i] = random.randint(-50,50)
    if(x2[i]-a*x1[i]-b>=0):
        y[i]=1
for i in range(10000):
    x1test[i] = random.randint(-50,50)
    x2test[i] = random.randint(-50,50)
    if(x2test[i]-a*x1test[i]-b>=0):
        ytest[i]=1
```

```python
fig = plt.figure()
ax = plt.axes()
plt.xlim(-50, 50)
plt.ylim(-50, 50)
x = np.linspace(-50, 50, 1000)
plt.plot(x,a*x+b)
for i in range(100):
    if(y[i]==1):
        plt.plot(x1[i],x2[i],'o',color='blue')
    else:
        plt.plot(x1[i],x2[i],'o',color='red')
plt.plot(x,-w[1]/w[2]*x-w[0]/w[2])
```

```python
def evaluate(x1,x2,w,y,incorrect):
    incorrect = []
    for i in range(100):
        output = np.sign(w[0]+w[1]*x1[i]+w[2]*x2[i]);
        if(y[i]!=output):
            incorrect.append(i)
    return incorrect
```

```python
incorrect = evaluate(x1,x2,w,y,incorrect)
count = 0
plt.xlim(-50, 50)
plt.ylim(-50, 50)
while(len(incorrect)>0 and count<=1000):
    index = random.choice(incorrect)
    signal = w[0]
    w[0] = w[0] + nu*(y[index]-signal)
    w[1] = w[1] + nu*(y[index]-signal)*x1[index]
    w[2] = w[2] + nu*(y[index]-signal)*x2[index]
    incorrect = evaluate(x1,x2,w,y,incorrect)
    count=count+1
    print("After ",count, "iterations")
    if(w[2]!=0):
        plt.plot(x,-w[1]/w[2]*x-w[0]/w[2])
```

```python
print("Final w:",w)
print("Final hypothesis: x2 = ",-w[1]/w[2],"x1 -",w[0]/w[2])
plt.xlim(-50, 50)
plt.ylim(-50, 50)
plt.plot(x,a*x+b)
for i in range(100):
    if(y[i]==1):
        plt.plot(x1[i],x2[i],'o',color='blue')
    else:
        plt.plot(x1[i],x2[i],'o',color='red')
plt.plot(x,-w[1]/w[2]*x-w[0]/w[2])
```

```python
def evaluatetest(x1,x2,w,y,incorrect):
    incorrect = []
    for i in range(10000):
        output = np.sign(w[0]+w[1]*x1[i]+w[2]*x2[i]);
        if(y[i]!=output):
            incorrect.append(i)
    return incorrect
```

```python
plt.xlim(-50, 50)
plt.ylim(-50, 50)
plt.plot(x,a*x+b)
for i in range(100):
    if(y[i]==1):
        plt.plot(x1[i],x2[i],'o',color='blue')
    else:
        plt.plot(x1[i],x2[i],'o',color='red')
plt.plot(x,-w[1]/w[2]*x-w[0]/w[2])
incorrect = evaluatetest(x1test,x2test,w,ytest,incorrect)
print("Test Error: ",len(incorrect))
```
