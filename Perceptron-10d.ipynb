```python
import random
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
a = [0] * 11
x = np.zeros((10,1000))
final_counts = [0]*100
y = [-1]*1000
w = [1]*11
for i in range(11):
    a[i] = random.randint(-10,10)
incorrect = []
print("Target function:",a[0],"x1 + ",a[1],"x2 + ",a[2],"x3 + ",a[3],"x4 + ",a[4],"x5 + ",a[5],"x6 + ",a[6],"x7 + ",a[7],"x8 + ",a[8],"x9 + ",a[9],"x10 + ",a[10],">=0")

for i in range(1000):
    signal = 0
    for j in range(10):
        x[j][i]=random.randint(-50,50)
        signal += a[j]*x[j][i]
    signal+=a[10]
    if(signal>=0):
        y[i]=1
print(y)
    
```

```python
def evaluate(x,w,y,incorrect):
    incorrect = []
    for i in range(1000):
        signal = 0
        for j in range(10):
            signal += w[j]*x[j][i]
        signal+=w[10]
        output = np.sign(signal);
        if(y[i]!=output):
            incorrect.append(i)
    return incorrect
```

```python
incorrect = evaluate(x,w,y,incorrect)
count = 0
while(len(incorrect)>0):
    index = random.choice(incorrect)
    for i in range(10):
        w[i] = w[i] + y[index]*x[i][index]
    w[10]=w[10]+y[index]
    incorrect = evaluate(x,w,y,incorrect)
    count=count+1
    print("After ",count, "iterations")
    print(len(incorrect))
```

```python
print("Total iterations: ",count)
print("Final w:",w)
print("Final hypothesis:",a[0],"x1 + ",w[1],"x2 + ",w[2],"x3 + ",w[3],"x4 + ",w[4],"x5 + ",w[5],"x6 + ",w[6],"x7 + ",w[7],"x8 + ",w[8],"x9 + ",w[9],"x10 + ",w[10],">=0")
```

```python
final_counts[0]=count
for case in range(99):
    w = [1]*11
    incorrect = evaluate(x,w,y,incorrect)
    count = 0
    while(len(incorrect)>0):
        index = random.choice(incorrect)
        for i in range(10):
            w[i] = w[i] + y[index]*x[i][index]
        w[10]=w[10]+y[index]
        incorrect = evaluate(x,w,y,incorrect)
        count=count+1
    print("Run ",case,": ",count)
    final_counts[case+1]=count
```

```python
plt.hist(final_counts,bins='auto')
plt.title("Number of iterations to convergence")
plt.show
```
